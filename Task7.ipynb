```python
#Q1
import seaborn as sns

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt



data = pd.read_csv('titanic.csv')


passenger_ids = data['PassengerId']
fares = data['Fare']


plt.figure(figsize=(20, 5))
plt.plot(passenger_ids, fares, marker='o', linestyle='-', color='r')


plt.title('Fare Against Passenger IDs')
plt.xlabel('Passenger ID')
plt.ylabel('Fare')
plt.xticks(ticks=range(1, len(passenger_ids) + 1, 50))

plt.grid(True)


plt.show()
```

```python

#Q2
pclass = data['Pclass'].value_counts()


plt.figure(figsize=(10, 6))
plt.bar(pclass.index, pclass.values, color='b')


plt.title('Frequency of Passengers in Different Classes')
plt.xlabel('Passenger Class')
plt.ylabel('Frequency')

plt.show()
```

```python
#Q3

data = data.dropna(subset=['Age', 'Fare'])


age = data['Age']
fare = data['Fare']


plt.figure(figsize=(10, 4))
plt.scatter(age, fare, alpha=0.5, c='blue', edgecolors='w', linewidth=0.5)


plt.title('Relationship between Age and Fare')
plt.xlabel('Age')
plt.ylabel('Fare')


plt.grid(True)


plt.show()
```

```python
#Q4
data = data.dropna(subset=['Age', 'Fare', 'Pclass', 'SibSp', 'Parch', 'Survived'])


selected_columns = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']
data_subset = data[selected_columns]


sns.pairplot(data_subset, hue='Survived')


plt.show()
```

```python
#Q5

plt.figure(figsize=(10, 6))
sns.boxplot(x='Pclass', y='Fare', data=data)


plt.title('Box Plot of Fares by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare')


plt.show()
```

```python
#Q6
c = ['Survived', 'Pclass', 'Age', 'SibSp', 'Parch', 'Fare']


subset_data = data[c]


corr_matrix = subset_data.corr()


plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Matrix Heatmap for Selected Columns')


plt.show()
```

```python
#Q7


titanic = sns.load_dataset('titanic')


fig, axs = plt.subplots(2, 2, figsize=(12, 10))


sns.countplot(x='survived', data=titanic, ax=axs[0, 0])

axs[0, 0].set_title('Survival Count')


sns.histplot(x='age', data=titanic, ax=axs[0, 1], kde=True)
axs[0, 1].set_title('Age Distribution')

sns.histplot(x='fare', data=titanic, ax=axs[1, 0], kde=True)
axs[1, 0].set_title('Fare Distribution')
sns.countplot(x='class', hue='survived', data=titanic, ax=axs[1, 1])
axs[1, 1].set_title('Survival by Class')


plt.tight_layout()
plt.show()

```

```python
#Q9

plt.figure(figsize=(10, 6))
sns.violinplot(x='class', y='age', hue='survived', data=titanic, split=True, inner='quart', palette='Set2')
plt.title('Distribution of Age by Passenger Class and Survival')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.legend(title='Survived', loc='upper left', labels=['No', 'Yes'])
plt.tight_layout()


plt.show()
```

```python
#Q10
fig, ax = plt.subplots(figsize=(8, 6))


sns.histplot(x='age', data=titanic, kde=True, ax=ax)


ax.set_xlabel('Age')

ax.set_ylabel('Frequency')

ax.set_title('Histogram with KDE for Age Distribution')


plt.show()
```
